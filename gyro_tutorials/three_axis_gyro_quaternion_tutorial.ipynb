# üõ∞Ô∏è Tutorial: 3-Axis Gyro Control and Quaternions for Attitude Representation

In this notebook, we demonstrate how a spacecraft with 3 orthogonal reaction wheels (gyros) uses **quaternions** to represent and control its orientation.

We'll:
- Simulate spacecraft rotation using applied torques on each axis
- Use quaternions to track orientation over time
- Convert quaternion results into Euler angles for visualization


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R

# Constants
I = np.diag([10.0, 12.0, 15.0])  # Inertia matrix (kg¬∑m¬≤)
invI = np.linalg.inv(I)

dt = 0.1  # timestep (s)
t = np.arange(0, 50, dt)

# Define control torques over time (simple square pulses)
torques = np.zeros((len(t), 3))
torques[10:30, 0] = 0.05  # X-axis torque
torques[20:40, 1] = -0.03  # Y-axis torque
torques[5:15, 2] = 0.04  # Z-axis torque

# Initialize angular velocity and quaternion
omega = np.zeros((len(t), 3))
q = np.zeros((len(t), 4))
q[0] = [1, 0, 0, 0]  # identity quaternion

for i in range(1, len(t)):
    # Euler's rotational equation: dœâ/dt = I‚Åª¬π (T - œâ√ó(Iœâ))
    w = omega[i-1]
    torque = torques[i]
    domega = invI @ (torque - np.cross(w, I @ w))
    omega[i] = omega[i-1] + domega * dt

    # Quaternion update (Rodrigues' rotation formula approximation)
    rotvec = omega[i] * dt  # angular velocity vector √ó dt
    dq = R.from_rotvec(rotvec).as_quat()  # [x, y, z, w]
    r_prev = R.from_quat(q[i-1])
    r_new = r_prev * R.from_quat(dq)
    q[i] = r_new.as_quat()

# Convert quaternions to Euler angles for plotting (ZYX order)
euler = R.from_quat(q).as_euler('zyx', degrees=True)
```

## üìä Plot Euler Angles Over Time

```python
plt.figure(figsize=(10,6))
plt.plot(t, euler[:, 0], label='Yaw (Z)')
plt.plot(t, euler[:, 1], label='Pitch (Y)')
plt.plot(t, euler[:, 2], label='Roll (X)')
plt.title('Orientation Over Time via Quaternion Integration')
plt.xlabel('Time (s)')
plt.ylabel('Angle (degrees)')
plt.legend()
plt.grid(True)
plt.show()
```

## ‚úÖ Summary

- Reaction wheels apply torque along body axes.
- The spacecraft reacts with rotation due to conservation of momentum.
- **Quaternions** are used to track 3D attitude without gimbal lock.
- We integrated the attitude over time and visualized the change in yaw, pitch, and roll.

This is the foundation of modern spacecraft **attitude determination and control systems (ADCS)**.
